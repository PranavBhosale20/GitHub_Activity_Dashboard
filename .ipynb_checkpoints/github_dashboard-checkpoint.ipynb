```python
# Step 2: Get and display GitHub user profile data

from utils.github_api_handler import get_user_profile

# Ask user for GitHub username
username = input("Enter GitHub username: ")

# Fetch profile data
profile_data = get_user_profile(username)

# Display key profile information
if profile_data:
    print(f"üë§ Name: {profile_data.get('name')}")
    print(f"üìç Location: {profile_data.get('location')}")
    print(f"üíº Bio: {profile_data.get('bio')}")
    print(f"üìä Public Repos: {profile_data.get('public_repos')}")
    print(f"üë• Followers: {profile_data.get('followers')}")
    print(f"üîó GitHub Profile: {profile_data.get('html_url')}")
else:
    print("‚ùå Could not fetch user profile.")
```

```python
# Step 3: Get and display repository data

repos_data = get_user_repos(username)

# Display summary
if repos_data:
    print(f"\nüì¶ Total Public Repositories: {len(repos_data)}")
    print("üóÉÔ∏è Repo Summary:")
    
    for repo in repos_data:
        print(f"\nüîπ Name: {repo['name']}")
        print(f"   ‚≠ê Stars: {repo['stargazers_count']}")
        print(f"   üç¥ Forks: {repo['forks_count']}")
        print(f"   üìù Language: {repo['language']}")
        print(f"   üîó URL: {repo['html_url']}")
else:
    print("‚ùå Could not fetch repositories.")
```

```python
# Step 4.1: Top 5 repositories by star count

# Sort by stars and take top 5
top_starred = sorted(repos_data, key=lambda x: x['stargazers_count'], reverse=True)[:5]

repo_names = [repo['name'] for repo in top_starred]
star_counts = [repo['stargazers_count'] for repo in top_starred]

# Plot
plt.figure(figsize=(8, 5))
sns.barplot(x=star_counts, y=repo_names, palette='viridis')
plt.title("‚≠ê Top 5 Repositories by Stars")
plt.xlabel("Stars")
plt.ylabel("Repository")
plt.tight_layout()
plt.show()
```

```python
# Step 4.2: Languages used

from collections import Counter

languages = [repo['language'] for repo in repos_data if repo['language']]
lang_counts = Counter(languages)

# Plot
plt.figure(figsize=(6, 6))
plt.pie(lang_counts.values(), labels=lang_counts.keys(), autopct='%1.1f%%', startangle=140)
plt.title("üìù Languages Used in Public Repositories")
plt.axis('equal')
plt.tight_layout()
plt.show()
```
